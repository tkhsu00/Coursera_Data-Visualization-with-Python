```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
df=pd.read_csv("Topic_Survey_Assignment.csv")
df.set_index('Unnamed: 0',inplace=True)
df
```

```python
df = df.sort_values(by=['Very interested'],ascending=False)
df
```

```python
cols = df.columns.tolist()
df['total'] = df.sum(axis=1)
df
```

```python
for col in df.columns:
    df[col] = df[col]/df['total']
    #df[col] = df[col].apply(lambda x: "{0:.2f}%".format(x*100))
df = df[cols]
```

```python
colors = ['#5cb85c','#5bc0de','#d9534f']
ax = df.plot(kind='bar', figsize=(20, 8), fontsize = 14, width=0.8, color = colors)

ax.set_title("Percentage of Respondents' Interest in Data Science Area", fontsize = 16)
ax.legend(fontsize = 14)

ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.axes.get_yaxis().set_ticks([])

ax.set_xlabel('')

for p in ax.patches:
    ax.annotate("{0:.2f}%".format(p.get_height()*100),(p.get_x(), p.get_height() * 1.03), fontsize = 14)
```
